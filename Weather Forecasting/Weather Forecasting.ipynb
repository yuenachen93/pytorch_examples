```python
# One-hot encoding
features = pd.get_dummies(features)
features.head(5)
```

```python
# labels
labels = np.array(features['actual'])

# Drop the labels from features
features = features.drop('actual', axis = 1)

# Save feature name
feature_list = list(features.columns)

# Transfer to suitable datatype
features = np.array(features)
```

```python
features.shape
```

```python
# Standarize data in different dimensions.
from sklearn import preprocessing
input_features = preprocessing.StandardScaler().fit_transform(features)
```

```python
input_features
```

### Construct Neural network Model

```python
x = torch.tensor(input_features, dtype = float)

y = torch.tensor(labels, dtype = float)

#Initialize Parameters
weights = torch.randn((14, 128), dtype = float, requires_grad = True)
biases = torch.randn(128, dtype = float, requires_grad = True)
weights2 = torch.randn((128, 1), dtype = float, requires_grad = True)
biases2 = torch.randn(1, dtype = float, requires_grad = True)

learning_rate = 0.001
losses = []

for i in range(1000):
    #calculate hidden layer
    hidden = x.mm(weights) + biases
    #add activate function
    hidden = torch.relu(hidden)
    #predict 
    predictions = hidden.mm(weights2) + biases2
    #calculate losses
    loss = torch.mean((predictions - y) ** 2)
    losses.append(loss.data.numpy())
    
    #print losses
    if i % 100 == 0:
        print('loss:', loss)
    #backward propagation
    loss.backward()
    
    #updata parameters
    weights.data.add_(- learning_rate * weights.grad.data)  
    biases.data.add_(- learning_rate * biases.grad.data)
    weights2.data.add_(- learning_rate * weights2.grad.data)
    biases2.data.add_(- learning_rate * biases2.grad.data)
    
    # clear gradient after each iteration
    weights.grad.data.zero_()
    biases.grad.data.zero_()
    weights2.grad.data.zero_()
    biases2.grad.data.zero_()
    
```

```python
predictions.shape
```

### Much simplier Net model

```python
input_size = input_features.shape[1] #number of features
hidden_size = 128
output_size = 1
batch_size = 16
my_nn = torch.nn.Sequential(
    torch.nn.Linear(input_size, hidden_size),
    torch.nn.Sigmoid(),
    torch.nn.Linear(hidden_size, output_size)
)

cost = torch.nn.MSELoss(reduction='mean') #loss function
optimizer = torch.optim.Adam(my_nn.parameters(), lr=0.001) #update parameters
```

```python
#Train
losses = []
for i in range(1000):
    batch_loss = []
    # Use batch
    for start in range(0, len(input_features), batch_size):
        end = start + batch_size if start + batch_size < len(input_features) else len(input_features)
        xx = torch.tensor(input_features[start:end], dtype = torch.float, requires_grad = True)
        yy = torch.tensor(labels[start:end], dtype = torch.float, requires_grad = True)
        prediction = my_nn(xx)
        loss = cost(prediction, yy)
        optimizer.zero_grad()
        loss.backward(retain_graph=True)
        optimizer.step()
        batch_loss.append(loss.data.numpy())
    #print loss
    if i%100 == 0:
        losses.append(np.mean(batch_loss))
        print(i, np.mean(batch_loss))
```

Prediction result

```python
x = torch.tensor(input_features, dtype = torch.float)
predict = my_nn(x).data.numpy()
```

```python
# Transfer to datetime frame 
dates = [str(int(year)) + '-' + str(int(month)) + '-' + str(int(day)) for year, month, day in zip(years, months, days)]
dates = [datetime.datetime.strptime(date, '%Y-%m-%d') for date in dates]

# Cretate a table to store datetime and corresponding labels 
true_data = pd.DataFrame(data = {'date': dates, 'actual': labels})

# Create a datatime and corresponding prediction result
months = features[:, feature_list.index('month')]
days = features[:, feature_list.index('day')]
years = features[:, feature_list.index('year')]

test_dates = [str(int(year)) + '-' + str(int(month)) + '-' + str(int(day)) for year, month, day in zip(years, months, days)]

test_dates = [datetime.datetime.strptime(date, '%Y-%m-%d') for date in test_dates]

predictions_data = pd.DataFrame(data = {'date': test_dates, 'prediction': predict.reshape(-1)}) 
```
